```python
import matplotlib.pyplot as plt
import numpy as np
from math import exp, log
%matplotlib inline
```

## Sigmoid:

The sigmoid function is the most common smooth step function. It is useful primarily for classification type problems.

$S(x) = \frac{1}{1+e^{-x}}=\frac{e^x}{e^x+1}$

$S'(x) = \frac{e^x}{(1+e^x)^2} = S(x)(1-S(x))$

```python
def S(x):
    return 1 / (1 + exp(-x))

X = np.arange(-6,6,.01)
Y = np.array([S(x) for x in X])
Z = Y*(1-Y)

fig = plt.figure(figsize=(15,6))
ax1 = plt.subplot(121)
ax1.plot(X,Y,label="$S(x)$")
ax2 = plt.subplot(122)
ax2.plot(X,Z,label="$S'(x)$")
ax1.plot([-6,6],[0,0],':k')
ax1.plot([0,0],[0,1],':k')
ax2.plot([-6,6],[0,0],':k')
ax2.plot([0,0],[0,0.25],':k')
ax1.set_title('$S(x)$')
ax2.set_title('$S\'(x)$')
plt.suptitle("Sigmoid Activation")
plt.show()
```

## Leaky ReLU

The ReLU (rectified linear unit) function is as follows:
$$
R(x) = \left\{
        \begin{array}{ll}
            0 & \quad x < 0 \\
            x & \quad x \geq 0
        \end{array}
    \right.
$$

```python
def R(x):
    return max(0,x)

def R_prime(x):
    if x >= 0:
        return 1
    else:
        return 0

X = np.arange(-6,6,.01)
Y = np.array([R(x) for x in X])
Z = np.array([R_prime(x) for x in X])

fig = plt.figure(figsize=(15,6))
ax1 = plt.subplot(121)
ax1.plot(X,Y,label="$R(x)$")
ax2 = plt.subplot(122)
ax2.plot(X,Z,label="$R'(x)$")
ax1.plot([-6,6],[0,0],':k')
ax1.plot([0,0],[0,6],':k')
ax2.plot([-6,6],[0,0],':k')
ax2.plot([0,0],[0,1],':k')
ax1.set_title('$R(x)$')
ax2.set_title('$R\'(x)$')
plt.suptitle("ReLU")
plt.show()
```

It should be noted that the derivative is undefined at 0; it is necessary to fill this value with the derivative from either side.

Of course, this function isn't viable as an activation function without modification; the zero slope on the left leads to stagnant weights; this is a problem which must be avoided. It is also wise to limit the size of the output. Here we define an activation function and a corresponding rate of change; this rate of change is not actually the derivative of the function in question; it includes nonzero slope where the function is constant, in order to avoid stagnant weights in neural nets. This innacurate nonzero slope is the "leakiness" in Leaky ReLU.

$$
LR(x,l,b) = \left\{
        \begin{array}{lll}
            0 & \quad x < 0 \\
            x & \quad 0\leq x\leq b \\
            b & \quad x > b
        \end{array}
    \right.
$$

$$
LR'(x,l,b) = \left\{
        \begin{array}{lll}
            l & \quad x < 0 \\
            1 & \quad 0\leq x\leq b \\
            l & \quad x > b
        \end{array}
    \right.
$$

```python
def LR(x,l,b):
    if x > b:
        return b
    elif x >= 0:
        return x
    else:
        return 0

def LR_prime(x,l,b):
    if x > b:
        return l
    elif x >= 0:
        return 1
    else:
        return l

l = 0.05
b = 5
X = np.arange(-6,6,.01)
Y = np.array([LR(x,l,b) for x in X])
Z = np.array([LR_prime(x,l,b) for x in X])

fig = plt.figure(figsize=(15,6))
ax1 = plt.subplot(121)
ax1.plot(X,Y,label="$LR(x,"+str(l)+","+str(b)+")$")
ax2 = plt.subplot(122)
ax2.plot(X,Z,label="$LR'(x,"+str(l)+","+str(b)+")$")
ax1.plot([-6,6],[0,0],':k')
ax1.plot([0,0],[0,6],':k')
ax2.plot([-6,6],[0,0],':k')
ax2.plot([0,0],[0,1],':k')
ax1.set_title("$LR(x,"+str(l)+","+str(b)+")$")
ax2.set_title("$LR\'(x,"+str(l)+","+str(b)+")$")
plt.suptitle("Leaky ReLU")
plt.show()
```

The Leaky ReLU function can predict linear functions perfectly, and can predict arbitrary nonlinear functions reasonably well given a large enough network.

## Logarithmic and Exponential Activation

I have devised some custom activation functions which behave like the natural log and like the exponential function within specified bounds, while maintaining the properties necessary for activations. The motivation for these two activations linked; a logarithmic layer whose output feeds into an exponential layer allows for arbitrary products:

$$
e^{a_2ln(a_1x) + b_2ln(b_1y)} = e^{ln((a_1x)^{a_2})}\cdot e^{ln((b_1y)^{b_2})} = (a_1x)^{a_2}(b_1y)^{b_2}
$$

I first define a smooth activator which behaves like the natural log when its input is between 1 and some upper bound $b$:

$$
LOG(x,b) = \left\{
        \begin{array}{lll}
            exp(x-1) & \quad x < 1 \\
            ln(x) + 1 & \quad 0\leq x\leq b \\
            ln(b) + 2 - e^{\frac{-x}{b}+1} & \quad x > b
        \end{array}
    \right.
$$

In practice, the bound $b$ for both the logarithmic and exponential activators should be an upper bound for the size of any realistic inputs which should be multiplied.

```python
def LOG(x,b):
    if x > b:
        return log(b) + 2 - exp(-x/b + 1)
    elif x >= 1:
        return log(x) + 1
    else:
        return exp(x-1)

def LOG_prime(x,b):
    if x > b:
        return (log(b) + 2 - LA(x,b)) / b
    elif x >= 1:
        return 1 / x
    else:
        return LA(x,b)

b = 10
X = np.arange(-6,20,.01)
Y = np.array([LOG(x,b) for x in X])
Z = np.array([LOG_prime(x,b) for x in X])

fig = plt.figure(figsize=(15,6))
ax1 = plt.subplot(121)
ax1.plot(X,Y,label="$LOG(x,"+str(b)+")$")
ax2 = plt.subplot(122)
ax2.plot(X,Z,label="$LOG'(x,"+str(b)+")$")
ax1.plot([-6,20],[0,0],':k')
ax1.plot([0,0],[0,6],':k')
ax2.plot([-6,20],[0,0],':k')
ax2.plot([0,0],[0,1],':k')
ax1.set_title("$LOG(x,"+str(b)+")$")
ax2.set_title("$LOG\'(x,"+str(b)+")$")
plt.suptitle("Logarithmic Activation")
plt.show()
```

Next, a smooth exponential activator; it behaves like an exponential function up to $x=ln(b)$ for some bound $b$, then exhibits exponential decay toward $y=2$:

$$
EXP(x,b) = \left\{
        \begin{array}{ll}
            2 - e^{-x+ln(b)} & \quad x > ln(b) \\
            \frac{e^x}{b} & \quad x \leq ln(b)
        \end{array}
    \right.
$$

```python
def EXP(x,b):
    if x > log(b):
        return 2 - exp(-x + log(b))
    else:
        return exp(x)/b

def EXP_prime(x,b):
    if x > log(b):
        return 2 - EXP(x,b)
    else:
        return EXP(x,b)

b = 10
X = np.arange(-6,10,.01)
Y = np.array([EXP(x,b) for x in X])
Z = np.array([EXP_prime(x,b) for x in X])

fig = plt.figure(figsize=(15,6))
ax1 = plt.subplot(121)
ax1.plot(X,Y,label="$EXP(x,"+str(b)+")$")
ax2 = plt.subplot(122)
ax2.plot(X,Z,label="$EXP'(x,"+str(b)+")$")
ax1.plot([-6,10],[0,0],':k')
ax1.plot([0,0],[0,2],':k')
ax2.plot([-6,10],[0,0],':k')
ax2.plot([0,0],[0,1],':k')
ax1.set_title("$EXP(x,"+str(b)+")$")
ax2.set_title("$EXP\'(x,"+str(b)+")$")
plt.suptitle("Exponential Activation")
plt.show()
```
